```python
import pandas as pd
from steamwebapi.api import ISteamUser, IPlayerService, ISteamUserStats
df1 = pd.DataFrame(columns=('steamId', 'owned_games', 'recently_games'))
f2 = open("steamid2.txt","r")
lines = f2.readlines()
i=0
for line3 in lines:
    steamuserinfo = IPlayerService(steam_api_key=key)
    owned_games=steamuserinfo.get_owned_games(line3)
    recently_games= steamuserinfo.get_recently_played_games(line3)
    df1 = df1.append([{'steamId':line3, 'owned_games':owned_games,'recently_games':recently_games }], ignore_index=True)
    i=i+1
    print(i)
    df1.to_csv('submissions_v2.csv',index=False, encoding='utf-8')
```
